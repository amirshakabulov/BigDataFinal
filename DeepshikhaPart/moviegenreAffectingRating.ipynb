```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

filename = open("/Users/deepshikha/Documents/spring 2023/BigData/FinalProject/BigDataFinal/Raw Data/IMDB Top 250 Movies.csv")
file = csv.DictReader(filename)
```

```python
dic={}
for col in file:
    if not col['genre'] in dic.keys():
        dic[col['genre']] = col['rating']
    dic[col['genre']] = str((float(dic[col['genre']]) + float(col['rating']))/2)    
```

```python
dicGenre = {}
for genre in dic.keys():
	ind = genre.split(',')
	for gen in ind:
		if not gen in dicGenre.keys():
			dicGenre[gen] = dic[genre]
		dicGenre[gen] = (float(dicGenre[gen]) + float(dic[genre]))/2
print(dicGenre)
```

```python
values = dicGenre.values()
labels = dicGenre.keys()

plt.bar(labels, values)

plt.title('Movie Ratings Affected by Genre')
plt.xlabel('Genres')
plt.ylabel('Ratings')
plt.ylim(7, 9)  # set the y-axis range
plt.xticks(rotation = 90)

plt.show()
```
